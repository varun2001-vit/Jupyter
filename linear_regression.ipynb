```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import pandas as pd
from torch.autograd import Variable
import torchvision
import torchvision.transforms as transforms
torch.set_printoptions(linewidth=120)
torch.set_grad_enabled(True)
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data=Variable(torch.tensor([[1.0],[2.0],[3.0]]))
y_data=Variable(torch.tensor([[2.0],[4.0],[6.0]]))
```

```python
class model(nn.Module):
    def __init__(self):
        super().__init__()
        #layer
        self.linear=nn.Linear(in_features=1,out_features=1)
    def forward(self,x):
        y_pred=self.linear(x)
        return y_pred
```

```python
mod=model()
```

```python
criterion=nn.MSELoss(size_average=False)
optimizer=optim.SGD(mod.parameters(),lr=0.01)
```

```python
for epoch in range(600):
    y_pred=mod(x_data)
    loss=criterion(y_pred,y_data)#passing the values
    print("epoch=",epoch," loss=",loss.item())
    optimizer.zero_grad() #setting to zero to remove cummalitve
    loss.backward() #updating gradients
    optimizer.step() #moving forward
```

```python
#after training
test=Variable(torch.tensor([[4.0]]))
print("value after training the model for value 4 is",mod.forward(test).data[0][0])
```
